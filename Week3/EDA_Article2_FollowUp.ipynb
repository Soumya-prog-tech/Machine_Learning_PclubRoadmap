It is showing that we don't have any null values in our datset but we have observed earlier that there were '?' symbols in the dataset, which means that these symbols are in the form of an object. Let us now chekc the data types of each attribute

```python
auto.info()
```

Checking for wrong entries like symbols -,?,#,* etc

```python
for col in auto.columns:
    print('{}:{}'.format(col, auto[col].unique()))
```

```python
for col in auto.columns : 
    auto[col] = auto[col].replace({'?':np.nan})
```

```python
auto.head()
```

```python
auto.isnull().sum()
```

```python
sns.heatmap(auto.isnull(), cbar = False, cmap = 'viridis')
plt.show()
```

```python
num_cols = auto.columns[auto.isnull().sum() > 0]
print(num_cols)
num_cols = num_cols.drop('num-of-doors')
num_cols
```

```python
for col in num_cols:
    auto[col] = pd.to_numeric(auto[col])
    auto[col] = auto[col].fillna(auto[col].mean())

auto.head()
```

```python
auto['num-of-doors'] = auto['num-of-doors'].fillna(auto['num-of-doors'].mode()[0])
```

```python
auto['num-of-doors'].unique()
```

```python
auto.head()
```

```python
auto_numeric = auto.select_dtypes(include = ['int64', 'float64'])
plt.figure(figsize = (10,10))
sns.heatmap(auto_numeric.corr(), cbar = True ,annot = True, cmap = 'YlGnBu')
plt.show()
```
